# Merging dataframes

```python
# We can merge the two dataframes on user_id variable
df_merge = ords_prods_merge.merge(customers, on = 'user_id', indicator = True)
print(df_merge.columns)
```

```python
# Check for the full match between both dataframes
df_merge['_merge'].value_counts()
```

```python
df_merge.head(10)
```

```python
# Check for mixed data types
for col in df_merge.columns.tolist():
  weird = (df_merge[[col]].map(type) != df_merge[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_merge[weird]) > 0:
    print (col)

# Find missing values
print('Missing values in merged dataframe: ' ,df_merge.isnull().sum())
```
