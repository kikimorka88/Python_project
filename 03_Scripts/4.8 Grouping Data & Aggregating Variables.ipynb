## Spending habits by loyalty status

```python
# Check the prices of the products ordered within each loyalty category
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean','min','max']})
```

```python
# Check the distribution of different price ranges among customer types
price_range_distribution = pd.crosstab(ords_prods_merge['loyalty_flag'], ords_prods_merge['price_range'])
print(price_range_distribution)
```

# Data check - prices

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Create the histogram of prices
ords_prods_merge['prices'].plot.hist(bins = 25)
plt.show()
print('Mean price', ords_prods_merge['prices'].mean())
print('Median price', ords_prods_merge['prices'].median())
print('Max price', ords_prods_merge['prices'].max())
# Create the scatter plot of prices
sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_merge)
plt.show()
print(ords_prods_merge.loc[ords_prods_merge['prices'] > 100])
# Exclude outliers
ords_prods_merge.loc[ords_prods_merge['prices'] >100, 'prices'] = np.nan
print('New max price', ords_prods_merge['prices'].max())
```

```python
# Visualize the normalized price range distribution to compare among customer types
price_range_distribution_normalized = price_range_distribution.div(price_range_distribution.sum(axis=1), axis=0) * 100
price_range_distribution_normalized.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.xlabel('Loyalty Flag')
plt.ylabel('Percentage of Purchases')
plt.title('Normalized Price Range Distribution within Each Customer Category')
plt.legend(title='Price Range', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig(os.path.join(path, '04_Analysis','Visualizations', 'bar_price-range_customer-types.png')) # Save the plot
```

#### From the basic stats we can see that the prices do not differ between the customer types. After closer inspection using the price range flags form the previous exercise, we see that most of the purchases in all groups are among mid-range products and the least purchases in all groups are in high-range products. Therefore, it seems that there is no difference in spending habits.

## Determine spender types

```python
# Calculate the average price per user
ords_prods_merge['avg_price_user'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
print(ords_prods_merge[['user_id','avg_price_user']].head(1))

# Create flags for spender types
ords_prods_merge.loc[ords_prods_merge['avg_price_user'] > 10, 'spender_type'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_price_user'] < 10, 'spender_type'] = 'Low spender'
ords_prods_merge['spender_type'].value_counts(dropna=False)
```

```python
ords_prods_merge.head(5)
```

## Determine frequent customers

```python
# Calculate median for days_since_prior_order per each customer
ords_prods_merge['med_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
print(ords_prods_merge[['user_id','med_days_since_prior_order']].head(1))

# Create flags for (non)frequent customer
ords_prods_merge.loc[ords_prods_merge['med_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['med_days_since_prior_order'] <= 20) & (ords_prods_merge['max_order'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['med_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
