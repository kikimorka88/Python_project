```python
# Import new file
ord_prod_cust = pd.read_pickle(os.path.join(path, '02_Data','Prepared_Data', 'ords_prods_customers_merge.pkl'))
col_list =  ['order_id', 'user_id', 'order_number', 'order_day_of_week',
       'order_hour_of_day', 'product_id', 'product_name','prices', 'price_range',
       'busiest_period_of_day', 'loyalty_flag', 'avg_price_user',
       'spender_type', 'name','surname', 'gender', 'age', 'date_joined', 
       'n_dependants', 'fam_status', 'income'] # Since we are dealing with a huge table, let's keep only the columns we are going to work with
ord_prod_cust = ord_prod_cust[col_list]
```

# Histogram of hourly ordering

```python
# Create histogram of hours distribution when orders were made
hist = ord_prod_cust['order_hour_of_day'].plot.hist(
    bins=24, 
    color='purple', 
    edgecolor='black'  # Adding borders to each bar
)
# Add y-axis label
plt.ylabel('Frequency')
# Add x-axis label
plt.xlabel('Hour of Day')
# Add title
plt.title('Distribution of Orders by Hour of Day')
```

```python
# Export the histogram
hist.figure.savefig(os.path.join(path, '04_Analysis','Visualizations', 'hist_hours.png'))
```

### We clearly see that the peak hours for ordering are between 10 and 16. 23 in the evening till 6 am are the hours with least orders and during the rest of the day the orders frequency is moderate.

# Customer demographics

## Age vs. family size

```python
# Reuse the subset extracted in the previous step to create another subset containing information about customer age and family size
sub3 = sub03[['age', 'n_dependants']]
```

```python
# Plot family size against the age - from the new subset of data
line3 = sns.lineplot(data = sub3, x = 'age',y = 'n_dependants', color = 'purple')
# Add y-axis label
plt.ylabel('Family Size')
# Add x-axis label
plt.xlabel('Age')
# Add title 
plt.title('Family Size against Customer Age')
```

```python
# Export the lineplot
line3.figure.savefig(os.path.join(path, '04_Analysis','Visualizations', 'line_family_age_sub03.png'))
```

### There is no clear direction of family size connected to age. This means that across all ages the family size strongly fluctuates and there are no trends. Besides, it fluctuates with a very small marigins, therefore there is no big effect of age on the family size. We cannot conclude that certain age group is prone to order more due to larger family size.
